```python
# Import necessary libraries for data analysis and visualization
import numpy as np
import pandas as pd
import plotly.express as px
from textblob import TextBlob

df = pd.read_csv('netflix_titles.csv')
```

```python
df.shape
```

```python
# Get a glimpse of the first few rows of the DataFrame
df.head()
```

```python
# View the column names of the DataFrame
df.columns
```

```python
# Group the DataFrame 'df' by the 'rating' column
x = df.groupby(['rating'])

# Count the occurrences within each group
x = x.size().reset_index(name='counts')

# Print the resulting DataFrame 'x'
print(x)
```

```python
# Create a pie chart using Plotly Express
pieChart = px.pie(x, values='counts', names='rating', title='Distribution of Rating')

# Display the pie chart
pieChart.show()
```

```python
# Replace missing values in 'director' column with 'No Director Specified'
df['director'] = df['director'].fillna('No Director Specified')

# Replace missing values in 'cast' column with 'No Cast Specified'
df['cast'] = df['cast'].fillna('No Cast Specified')
```

```python
df.head()
```

```python
# Create a DataFrame from split director names
directors_list = pd.DataFrame(df['director'].str.split(',', expand=True).stack())

# Print the DataFrame 'directors_list'
print(directors_list)
```

```python
# Rename the single column in 'directors_list' to 'Directors'
directors_list.columns = ['Directors']

# Print the DataFrame with the renamed column
print(directors_list)
```

```python
# Group 'directors_list' by 'Directors' and count occurrences
directors = directors_list.groupby('Directors').size().reset_index(name='Total Counts')

# Print the DataFrame 'directors' containing director counts
print(directors)
```

```python
# Filter 'directors' DataFrame to exclude 'No Director Specified'
directors = directors[directors.Directors != 'No Director Specified']

# Print the filtered DataFrame 'directors'
print(directors)
```

```python
# Sort 'directors' DataFrame by 'Total Counts' in descending order
directors = directors.sort_values(by=['Total Counts'], ascending=False)

# Print the sorted DataFrame 'directors'
print(directors)
```

```python
# Get the top 5 directors (most frequent)
top5directors = directors.head()

# Print the DataFrame 'top5directors'
print(top5directors)
```

```python
# Sort 'top5directors' by 'Total Counts' in ascending order (optional)
# Since you previously sorted directors by total counts in descending order,
# this line is not strictly necessary here for the bar chart. However,
# uncommenting it would display directors in ascending order (least to most frequent).

# top5directors = top5directors.sort_values(by=['Total Counts'], ascending=True)

# Create a bar chart using Plotly Express for the top 5 directors
barChart = px.bar(top5directors, x='Total Counts', y='Directors', title='Top 5 Directors on Netflix')

# Display the bar chart
barChart.show()
```

```python
# Create a DataFrame 'cast_df' to hold individual cast members
cast_df = pd.DataFrame()
cast_df = df['cast'].str.split(',', expand=True).stack()
cast_df = pd.DataFrame(cast_df)

# Rename the single column in 'cast_df' to 'Actor'
cast_df.columns = ['Actor']

# Group 'cast_df' by 'Actor' and count occurrences
actors = cast_df.groupby(['Actor']).size().reset_index(name='Total Count')

# Filter 'actors' DataFrame to exclude 'No Cast Specified'
actors = actors[actors.Actor != 'No Cast Specified']

# Sort 'actors' DataFrame by 'Total Counts' in descending order (most frequent first)
actors = actors.sort_values(by=['Total Count'], ascending=False)

# Get the top 5 actors (most frequent)
Top5actors = actors.head()

# Sort 'Top5actors' DataFrame by 'Total Counts' in ascending order (optional)
# This line is commented out as you might prefer to keep the descending order.
# Top5actors = Top5actors.sort_values(by=['Total Count'], ascending=True)

# Create a bar chart using Plotly Express for the top 5 actors
barchart2 = px.bar(Top5actors, x='Total Count', y='Actor', title='Top 5 Actors on Netflix')

# Display the bar chart
barchart2.show()
```

```python
# Select specific columns 'type' and 'release_year' from DataFrame 'df'
df1 = df[['type', 'release_year']]

# Rename columns in 'df1' for better readability
df1 = df1.rename(columns={"release_year": "Release Year", "type": "Type"})

# Group data by 'Release Year' and 'Type', count occurrences, and reset index
df2 = df1.groupby(['Release Year', 'Type']).size().reset_index(name='Total Count')

# Filter data in 'df2' to include years 2000 and above
df2 = df2[df2['Release Year'] >= 2000]

# Create line chart using Plotly Express
graph = px.line(df2, x='Release Year', y='Total Count', color='Type', title='Trend of content produced on Netflix')

# Display the line chart
graph.show()
```

```python
df['country']=df['country'].fillna('No Country Specified')
```

```python
# Split the 'country' column into separate rows
df = df.assign(country=df['country'].str.split(',')).explode('country')

# Groupby and count
country = df.groupby('country').size().reset_index(name='Total Count')

# Exclude 'No Country Specified'
country = country[country['country'] != 'No Country Specified']

# Sort and get the top 5
country = country.sort_values(by='Total Count', ascending=False).head(5)

# Create the bar chart
barchart = px.bar(country, x='country', y='Total Count', title='Top 5 Countries')
barchart.show()
```

```python
#Sentimental Analysis of Netflix Content
# Select specific columns 'release_year' and 'description' from DataFrame 'df'
df3 = df[['release_year', 'description']]

# Rename columns in 'df3' for better readability
df3 = df3.rename(columns={"release_year": "Release Year", "description": "Description"})

# Add a new column 'Sentiment' to store sentiment analysis results
df3['Sentiment'] = None

# Loop through each row in 'df3' to analyze sentiment
for index, row in df3.iterrows():
    # Extract description text from the current row
    description = row['Description']

    # Use TextBlob to analyze sentiment polarity
    testimonial = TextBlob(description)
    polarity = testimonial.sentiment.polarity

    # Determine sentiment based on polarity score
    if polarity == 0:
        sentiment = 'Neutral'
    elif polarity > 0:
        sentiment = 'Positive'
    else:
        sentiment = 'Negative'

    # Update the 'Sentiment' column for the current row
    df3.loc[index, 'Sentiment'] = sentiment

# Group data by 'Release Year' and 'Sentiment', count occurrences, and reset index
df3 = df3.groupby(['Release Year', 'Sentiment']).size().reset_index(name='Total Count')

# Filter data in 'df3' to include years 2006 and above (excluding 2005)
df3 = df3[df3['Release Year'] > 2005]

# Create bar chart using Plotly Express
barGraph = px.bar(df3, x='Release Year', y='Total Count', color='Sentiment', title='Sentiment Analysis of Netflix Content')

# Display the bar chart
barGraph.show()
```
